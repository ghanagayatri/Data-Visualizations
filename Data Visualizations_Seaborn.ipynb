### Explore The Distribution of Data using Seaborn

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# shows the distribution of univariate data with KDE ( Kernel Density Distribution )
sns.distplot(tips['total_bill'])
```

```python
# without KDE
sns.distplot(tips['total_bill'],kde=False)
```

```python
# Set the number of bins to dive deeper into the data
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
# Comparing Bivariate Data
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# To look deeper into the data we can use the variable "Kind"
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Exploring Bivariate Data with a regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# Exploring Bivariate Data with KDE ( Shows where the density is its darker)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# Pariwise plot compares all the numerical valued columns across the entire dataset
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```
